```python
no_months = len(df['Month'])
amount_return = sum(df['Amount returned to everyone in the group'])
#bidder_amount = df['Amount won by the bidder'][0]
#bidder_net = df['Net amount recd by Bid winner'][0]
#total_inves = no_months*df['Contribution'][0]
#total_return = bidder_net
#cummlative_return = (total_return-(total_inves-amount_return))/(total_inves-amount_return)
#annualized_return = ((1+cummlative_return)**(12/no_months)) - 1
```

```python
x1 = []
x2 = []
x3 = []
x4 = []
for i in range(0,25):
    bidder_amount = df['Amount won by the bidder'][i]     #Amount won by bidder after without removing commission
    bidder_net = df['Net amount recd by Bid winner'][i]   #Amount won by bidder after with removing commission
    total_inves = no_months*df['Contribution'][i] #Total investment 50,000/-
    total_return = bidder_net # total return after removing commission
    cummlative_return1 = ((total_return-(total_inves-amount_return))/(total_inves-amount_return))#for total investments
    cummlative_return2 = ((total_return-(total_inves-amount_return))/(total_return))# for total returns
    annualized_return1 = (((1+cummlative_return1)**(12/no_months)) - 1)*100  # Returns of each month bidder total investment using Compound interest
    annualized_return2 = (((1+cummlative_return2)**(12/no_months)) - 1)*100 # Returns of each month bidder total returns using Compound interest
    annualized_return3 =  ((cummlative_return1*12)/no_months)*100 # Returns of each month bidder total investment using Simple interest
    annualized_return4 =  ((cummlative_return2*12)/no_months)*100 # Returns of each month bidder total returns using Simple interest
    x1.append(annualized_return1) 
    x2.append(annualized_return2)   # Storing all the values in four variables
    x3.append(annualized_return3)
    x4.append(annualized_return4)
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib notebook
N = 25
N1 = np.arange(N)
#Width = 10
plt.figure()
p1 = plt.plot(N1,x1,color = 'green')
p2 = plt.plot(N1,x2,color = 'darkblue')
p3 = plt.plot(N1,x3,color = 'red')
p4 = plt.plot(N1,x4,color = 'pink')
plt.legend( (p1[0], p2[0],p3[0], p4[0]),('case1', 'case2','case3','case4'))
plt.ylabel('values(returns)')
plt.xlabel('Months')
plt.title('Annualized return for all cases')
#plt.xticks(np.arange(0,25, 1))
```
